```python
# replace 0 with NaN in BUDGET
final.BUDGET = [np.nan if x  == 0 else x for x in final.BUDGET]
```

```python
# check if null value has other representatives in the columns
final.replace('None', np.nan, inplace=True)
final.replace('none', np.nan, inplace=True)
```

```python
# extract shows information, primary key: IMDB_ID
shows = final[['IMDB_ID','Title','Series or Movie', 'Runtime', 'View Rating', 'Release Date','BUDGET','Boxoffice']]
shows
```

```python
# extract show performance information

show_perf = final[['IMDB_ID', 'IMDb Score', 'IMDb Votes', 'Rotten Tomatoes Score', 'Metacritic Score', 'Awards Received', 'Awards Nominated For', 'POPULARITY', 'VOTE_AVERAGE', 'VOTE_COUNT']]
show_perf
```

```python
# extract col info --> primary key: col_id

def extract_col_info(show_col, col, col_id, col_name):
    df = []
    for i in tqdm(show_col.dropna()[col].values):
        for c in i.split(','):
            df.append(c.strip())
    unique_col = list(set(df))
    df_id = []
    for i in range(len(unique_col)):
        df_id.append(col[:2]+str(i+1))

    col_info = {col_id:df_id, col_name: unique_col}
    
    return pd.DataFrame.from_dict(col_info)



# create relation table between col_info and show_info (many -- many)

def create_relation_table(show_col, col_info, col_id, col, col_name):
    '''
    @show_col: dataframe
    @col_info: dataframe
    @col_id: string, one column name in col_info
    @col: string, one column name in show_col
    @col_name: string, one column name in col_info
    '''

    show_col_relation = pd.DataFrame(columns = ['IMDB_ID',col_id])
    for i in tqdm(show_col.dropna().index):
    #     if i > 2:
    #         print(i)
    #         break
        d = {'IMDB_ID': [],col_id: []}
        for c in show_col.loc[i,col].split(','):
            c = c.strip()
            d['IMDB_ID'].append(show_col.loc[i,'IMDB_ID'])
            d[col_id].append(col_info[col_info[col_name] == c][col_id].values[0])
        d = pd.DataFrame.from_dict(d)
        show_col_relation = pd.concat([show_col_relation,d],ignore_index=True)

#     empty = show_col[show_col[col].isna()].rename(columns = {col: col_id})
#     show_col_relation = pd.concat([show_col_relation, empty], ignore_index=True)
    
    return show_col_relation
```

```python
final.columns
```

```python
# show and genre
show_Genre = final[['IMDB_ID', 'Genre']]
Genres = extract_col_info(show_col=show_Genre, col='Genre', col_id='Genre_id', 
                             col_name='Genre_name')
show_Genre_relation = create_relation_table(show_col=show_Genre, col_info=Genres, 
                                           col_id='Genre_id', col='Genre', col_name='Genre_name')
print(f'Genre: {Genres.shape}')
print(f'show_Genre_relation: {show_Genre_relation.shape}\n')
```

```python
# show and Language
show_Language = final[['IMDB_ID', 'Languages']]
show_Language = show_Language.rename(columns={'Languages': 'Language'})
Languages = extract_col_info(show_col=show_Language, col='Language', col_id='Language_id', 
                             col_name='Language_name')
show_Language_relation = create_relation_table(show_col=show_Language, col_info=Languages, 
                                           col_id='Language_id', col='Language', col_name='Language_name')
print(f'Languages: {Languages.shape}')
print(f'show_Language_relation: {show_Language_relation.shape}\n')
```

```python
# show and Country
show_Country = final[['IMDB_ID', 'Country Availability']]
show_Country = show_Country.rename(columns={'Country Availability': 'Country'})
Countries = extract_col_info(show_col=show_Country, col='Country', col_id='Country_id', 
                             col_name='Country_name')
show_Country_relation = create_relation_table(show_col=show_Country, col_info=Countries, 
                                           col_id='Country_id', col='Country', col_name='Country_name')
print(f'Countries: {Countries.shape}')
print(f'show_Country_relation: {show_Country_relation.shape}\n')
```

```python
# show and Director
show_Director = final[['IMDB_ID', 'Director']]
# show_Director = show_Country.rename(columns={'Country Availability': 'Country'})
Directors = extract_col_info(show_col=show_Director, col='Director', col_id='Director_id', 
                             col_name='Director_name')
show_Director_relation = create_relation_table(show_col=show_Director, col_info=Directors, 
                                           col_id='Director_id', col='Director', col_name='Director_name')
print(f'Directors: {Directors.shape}')
print(f'show_Director_relation: {show_Director_relation.shape}\n')
```

```python
# show and Writer
show_Writer = final[['IMDB_ID', 'Writer']]
# show_Director = show_Country.rename(columns={'Country Availability': 'Country'})
Writers = extract_col_info(show_col=show_Writer, col='Writer', col_id='Writer_id', 
                             col_name='Writer_name')
show_Writer_relation = create_relation_table(show_col=show_Writer, col_info=Writers, 
                                           col_id='Writer_id', col='Writer', col_name='Writer_name')
print(f'Writers: {Writers.shape}')
print(f'show_Writer_relation: {show_Writer_relation.shape}\n')
```

```python
# show and Actor
show_Actor = final[['IMDB_ID', 'Actors']]
show_Actor = show_Actor.rename(columns={'Actors': 'Actor'})
Actors = extract_col_info(show_col=show_Actor, col='Actor', col_id='Actor_id', 
                             col_name='Actor_name')
show_Actor_relation = create_relation_table(show_col=show_Actor, col_info=Actors, 
                                           col_id='Actor_id', col='Actor', col_name='Actor_name')
print(f'Actors: {Actors.shape}')
print(f'show_Actor_relation: {show_Actor_relation.shape}\n')
```
